```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.feature_selection import SelectKBest, chi2
```

```python
X = data.drop(columns =['status'],axis = 1)
original_feature_names = X.columns.tolist()#storing original feature names
y = data['status']
urls = X['url'] # storing url separately
```

```python
original_feature_names
```

```python
url_transformer = CountVectorizer()
X_url = url_transformer.fit_transform(X['url'])
```

```python
categorical_features = [col for col in X.columns if col != 'url']
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
X_categorical = categorical_transformer.fit_transform(X[categorical_features])
```

```python
X_transformed = pd.concat([pd.DataFrame(X_url.toarray()), pd.DataFrame(X_categorical.toarray())], axis=1)
# Scale the features to a non-negative range
scaler = MinMaxScaler()
X_transformed_scaled = scaler.fit_transform(X_transformed)
```

```python
k = 60  # Number of features you want to select
selector = SelectKBest(score_func=chi2, k=k)
X_new = selector.fit_transform(X_transformed_scaled, y)

# Get the indices of the selected features
selected_features_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_features_names = X_transformed.columns[selected_features_indices]
```

```python
print("Selected features:")
print(selected_features_names)
```

```python
# trying to retrieve the column names
print("Shape of X_transformed:", X_transformed.shape)
print("Selected feature indices:", selected_features_indices)
```

```python
# Print original feature names
print("Original feature names:")
print(original_feature_names)

# Print selected feature indices
print("\nSelected feature indices:")
print(selected_features_names)

# Print the length of original feature names
print("\nNumber of original features:", len(original_feature_names))

# Check if any selected feature index exceeds the length of original feature names
invalid_indices = [idx for idx in selected_features_indices if idx >= len(original_feature_names)]
print("\nInvalid indices:", invalid_indices)
```

```python
# from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_selected = pd.DataFrame(X_transformed_scaled[:, selected_features_indices])#names > indices
```

```python
X_url_df = pd.DataFrame(X_url.toarray(), columns=url_transformer.get_feature_names_out())
new_dataset = pd.concat([urls, X_selected], axis=1)
# Optionally, include the target variable in the new dataset if needed
new_dataset['status'] = y
```

```python
new_dataset_file = 'selected_features_dataset.csv'
new_dataset.to_csv(new_dataset_file, index=False)
```
